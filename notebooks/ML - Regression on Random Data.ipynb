```python
import random
import numpy as np
import os
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
inputFeatures = np.random.randint(9, size=(1000, 4))
```

```python
def resolveFunc(rndArr):
    lst = []
    for nda in rndArr:
        a = nda[0]*2
        b = nda[1]*9
        c = nda[2]*8
        d = nda[3]*3
        result = 5+a+b+c+d
        lst.append(result)
    retArr = np.reshape(lst, (1000, 1))
    return retArr
```

```python
labelArr = resolveFunc(inputFeatures)
```

```python
if len(inputFeatures) == len(labelArr):
    print("good to go")
else:
    print("messed up")
```
